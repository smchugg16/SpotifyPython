```python
#importing libraries
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#importing dataset 
df_tracks = pd.read_csv('C:/Users/green/Downloads/Spotify Datasets/tracks.csv')
df_tracks.head()
```

```python
#count null values
pd.isnull(df_tracks).sum()
```

```python
#dataset information
df_tracks.info()
```

```python
#10 least popular tracks in dataset 
sorted_df = df_tracks.sort_values('popularity', ascending = True).head(10)
sorted_df
```

```python
#stats about tracks
df_tracks.describe().transpose()
```

```python
#top 10 most popular tracks in descending order
most_popular = df_tracks.query('popularity > 90', inplace = False).sort_values('popularity', ascending = False)
most_popular[:10]
```

```python
df_tracks.set_index("release_date", inplace = True)
df_tracks.index = pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
#changing duration format to seconds
df_tracks["duration"] = df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace = True, axis=1)
```

```python
df_tracks.duration.head()
```

```python
#creating correlation plots
corr_df = df_tracks.drop(["key","mode","explicit"],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g", vmin=-1, vmax=1, center=0, cmap="viridis", linewidths=1, linecolor="Black")
heatmap.set_title("Correlation HeatMap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
sample_df = df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x = "energy", color = "limegreen").set(title = "Loudness vs. Energy Correlation")
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x = "acousticness", color = "darkgreen").set(title = "Popularity vs. Acousticness Correlation")
```

```python
#release year of tracks
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
#number of songs per year available on spotify histogram
sns.displot(years,discrete=True, aspect=2, height=5, kind="hist").set(title="Number of Songs per Year Available on Spotify")
```

```python
#year vs. duration barchart
total_dr = df_tracks.duration
fig_dims = (18,7)
fig, ax = plt.subplots(figsize = fig_dims)
fig = sns.barplot(x=years, y=total_dr, ax=ax, errwidth = False).set(title="Year vs. Duration")
plt.xticks(rotation=90)
```

```python
#Year vs. Duration plot
total_dr = df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims= (10,5)
fig, ax = plt.subplots(figsize=fig_dims)
fig = sns.lineplot(x=years, y=total_dr, ax=ax).set(title="Year vs. Duration")
plt.xticks(rotation=60)
```
